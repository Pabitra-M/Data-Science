```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.0, 5.0, 100)
y = np.sin(x)
print (x)
print (y)

plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot Ex')
plt.grid(True)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([3,5,7,2,1])


plt.scatter(x, y, c='red', marker='o', edgecolors='black')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot Ex')
plt.grid(True)
plt.show()
```

```python
cat = ['A', 'B', 'C', 'D']
valu = np.random.randint(20,100,4)

plt.bar(cat,valu, color=['blue' , 'green' ,'red', 'orange'])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('bar Ex')
plt.xticks(rotation=45)
plt.show()
```
