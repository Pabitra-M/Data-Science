```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
x = np.random.normal(loc=50, scale=10, size=1000)



plt.hist(x, bins=20, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Ex')
plt.grid(True)
plt.show()
```

```python
data = np.random.rand(5,5)
print (data)

plt.imshow(data, cmap='viridis')
plt.colorbar(label = 'value')
```

```python
fruits = np.array(['Apple', 'Orange' , 'Banana'])
count= np.array([20,30,28])


plt.pie(count, labels=fruits, autopct='%1.1f%%')
plt.title('pie')
plt.show()
```

```python
sco = {'class A' : [60,64,68,80,85],
      'class B' : [40,44,48,50,55],
      'class C' : [20,24,28,30,35]}
plt.boxplot(sco.values(), labels=sco.keys())
plt.title("boxplot")
plt.show()
```

```python
sco = {'Tem' : [20,25,40,34,30],
      'hum' : [40,44,50,45,55],
      'groth' : [5,7,6,5,8]}

data = pd.DataFrame(sco)
sns.pairplot(data)
plt.show()

```
