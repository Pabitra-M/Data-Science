```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
dataset = pd.read_csv("Salary_Data.csv")
dataset.head()
```

```python
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.scatter(dataset.YearsExperience, dataset.Salary, color ='red', marker = 'o')
```

```python
x = dataset.drop("Salary", axis = "columns")
y = dataset.Salary
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
modal = LinearRegression()
modal.fit(x_train, y_train)
```

```python
modal.score(x_test, y_test)
```

```python
y_lo_pred = modal.predict(x_test)

```

```python
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.scatter(x_test, y_test, color ='red', marker = 'x')
plt.plot(x_test, y_lo_pred, color="g")
```
